# **Representa√ß√£o de Texto**

- **TF-IDF  e TF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stopwords_pt = stopwords.words('portuguese')

assert len(cleaned_messages) == len(regressao), "N√∫mero de mensagens n√£o coincide com os r√≥tulos"

def tf_representation(cleaned_messages):
    tf_vectorizer = TfidfVectorizer(stop_words=stopwords_pt, ngram_range=(1, 2), use_idf=False)  # TF only
    X_tf = tf_vectorizer.fit_transform(cleaned_messages)
    return X_tf, tf_vectorizer

def tfidf_representation(cleaned_messages):
    tfidf_vectorizer = TfidfVectorizer(stop_words=stopwords_pt, ngram_range=(1, 2), use_idf=True)  # TF-IDF
    X_tfidf = tfidf_vectorizer.fit_transform(cleaned_messages)
    return X_tfidf, tfidf_vectorizer

X_tf, tf_vectorizer = tf_representation(cleaned_messages)
X_tfidf, tfidf_vectorizer = tfidf_representation(cleaned_messages)

X_tf_test = tf_vectorizer.transform(cleaned_messages_test)
X_tfidf_test = tfidf_vectorizer.transform(cleaned_messages_test)
```

- **Embeddings**

```python
from gensim.models import Word2Vec
import numpy as np
```

```python
def word2vec_representation(cleaned_messages, vector_size=100, window=5, min_count=1):
    tokenized_sentences = [re.findall(r'\w+', message.lower()) for message in cleaned_messages]

    word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=vector_size, window=window, min_count=min_count, workers=4)

    def sentence_to_vector(sentence):
        word_vectors = [word2vec_model.wv[word] for word in sentence if word in word2vec_model.wv.key_to_index]
        if not word_vectors:
            return np.zeros(word2vec_model.vector_size)
        return np.mean(word_vectors, axis=0)

    X_word2vec = np.array([sentence_to_vector(sentence) for sentence in tokenized_sentences])
    return X_word2vec, word2vec_model

X_word2vec, word2vec_model = word2vec_representation(cleaned_messages)
X_word2vec_test, word2vec_model_test = word2vec_representation(cleaned_messages_test)
```

# **Gr√°fico comparativo**

```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor  
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.svm import SVR 


y = np.array(regressao)
y_test = np.array(regressao_test)

def calcular_metrica(y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    return mae, mse, rmse

def knn_model(X_train, y_train, X_test, y_test, n_neighbors=3):
    knn = KNeighborsRegressor(n_neighbors=n_neighbors)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return calcular_metrica(y_test, y_pred)

def decision_tree_model(X_train, y_train, X_test, y_test, max_depth=None):
    tree = DecisionTreeRegressor(max_depth=max_depth, random_state=42)
    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_test)
    return calcular_metrica(y_test, y_pred)

def random_forest_model(X_train, y_train, X_test, y_test, n_estimators=100, max_depth=None):
    rf = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    return calcular_metrica(y_test, y_pred)


def svm_model(X_train, y_train, X_test, y_test, kernel='rbf', C=1.0, epsilon=0.1):
    svm = SVR(kernel=kernel, C=C, epsilon=epsilon)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    return calcular_metrica(y_test, y_pred)

resultados = {
    "TF": {
        "KNN": knn_model(X_tf, y, X_tf_test, y_test),
        "√Årvore de Decis√£o": decision_tree_model(X_tf, y, X_tf_test, y_test),
        "Random Forest": random_forest_model(X_tf, y, X_tf_test, y_test),
        "SVM": svm_model(X_tf, y, X_tf_test, y_test)
    },
    "TF-IDF": {
        "KNN": knn_model(X_tfidf, y, X_tfidf_test, y_test),
        "√Årvore de Decis√£o": decision_tree_model(X_tfidf, y, X_tfidf_test, y_test),
        "Random Forest": random_forest_model(X_tfidf, y, X_tfidf_test, y_test),
        "SVM": svm_model(X_tfidf, y, X_tfidf_test, y_test)
    },
    "Word2Vec": {
        "KNN": knn_model(X_word2vec, y, X_word2vec_test, y_test),
        "√Årvore de Decis√£o": decision_tree_model(X_word2vec, y, X_word2vec_test, y_test),
        "Random Forest": random_forest_model(X_word2vec, y, X_word2vec_test, y_test),
        "SVM": svm_model(X_word2vec, y, X_word2vec_test, y_test)
    }
}

dados = []
metricas = ["MAE", "MSE", "RMSE"]

for representacao, modelos in resultados.items():
    for modelo, valores in modelos.items():
        for metrica, valor in zip(metricas, valores):
            dados.append({
                "Representa√ß√£o": representacao,
                "Modelo": modelo,
                "M√©trica": metrica,
                "Valor": round(valor, 4)
            })

df = pd.DataFrame(dados)

print(df)

fig = px.bar(
    df,
    x="M√©trica",
    y="Valor",
    color="Modelo",
    facet_col="Representa√ß√£o",
    barmode="group",
    text="Valor",
    title="Compara√ß√£o de Erros (MAE, MSE, RMSE) para Cada Modelo e Representa√ß√£o",
    labels={
        "Valor": "Valor de Erro",
        "M√©trica": "M√©trica de Erro",
        "Modelo": "Modelo",
        "Representa√ß√£o": "Representa√ß√£o"
    },
    height=600,
    width=900
)

fig.update_layout(
    uniformtext_minsize=8,
    uniformtext_mode='hide',
    legend_title_text='Modelo',
    title_x=0.5
)

fig.show()

```
