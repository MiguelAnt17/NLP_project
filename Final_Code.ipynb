# **Zero-Shot**

  - **zero-shot - temperatura = 0.1**

 - **Llama3.2 com 3B de par√¢metros**

```python
msg_model = df_secundario_test["Msg"][:400]
```

```python
from langchain_ollama import OllamaLLM

messages = msg_model

model = 'llama3.2:3b'  
temperature = 0.1  
results_llama3b_zero = [] 

llm = OllamaLLM(model=model, temperature=temperature)
    
for msg in messages:
    prompt = (
        f'Avalie o sentimento da mensagem fornecendo um n√∫mero entre -2 e 1. '
        f'Responda apenas com o n√∫mero, sem texto adicional. A mensagem √©: "{msg}"'
    )
    response = llm(prompt)
    response = response.strip().replace(',', '.')
    
    results_llama3b_zero.append(response)   
    print(f"Processado: Mensagem {len(results_llama3b_zero)}/{len(messages)}")
    print(results_llama3b_zero[-1])

print("Resultados Finais:")
for i, resultado in enumerate(results_llama3b_zero):
    print(f"Mensagem {i+1}: {resultado}")
```

# **Few-Shot**

 - usar os 10 primeiros tweets como exemplos

 - **Llama3.2 com 3B de par√¢metros**

```python
from langchain_ollama import OllamaLLM

messages = msg_model  

model = 'llama3.2:3b'  
temperature = 0.1  

results_llama3b_few = [] 

example_text = """\
Exemplo 1:
Mensagem: "al√¥, preciso de internet pelo menos, ajudem me."
Classifica√ß√£o: -1.0

Exemplo 2:
Mensagem: "Ol√° lamentamos a situa√ß√£o que nos reporta, mas existe uma anomalia a decorrer em algumas zonas do pa√≠s. No entanto, a nossa equipa t√©cnica j√° se encontra a retificar esta situa√ß√£o. Estamos ao seu dispor para ajudar. Obrigado e at√© breve!"
Classifica√ß√£o: 0.0

Exemplo 3:
Mensagem: "Posso terminar o contrato e mudar para outra operadora sem custos? J√° que voc√™s n√£o t√™m capacidade de fornecer um servi√ßo b√°sico."
Classifica√ß√£o: -1.5
"""

for i, msg in enumerate(messages):
    prompt = (
        f"Avalie o sentimento da mensagem fornecendo **apenas um n√∫mero** entre -2 e 1. "
        f"N√£o inclua texto adicional na resposta. Use os exemplos abaixo como refer√™ncia:\n\n{example_text}\n\n"
        f"Mensagem: \"{msg}\"\nClassifica√ß√£o:"
    )
    try:
        response = llm(prompt)
        response = response.strip()  
        results_llama3b_few.append(response)  
        print(f"Processado: Mensagem {i+1}/{len(messages)} - Classifica√ß√£o: {response}")
    except Exception as e:
        results_llama3b_few.append(f"Erro: {str(e)}")
        print(f"Erro ao processar mensagem {i+1}: {e}")

print("\nResultados Finais:")
for i, resultado in enumerate(results_llama3b_few):
    print(f"Mensagem {i+1}: {resultado}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

final_results_without_msg = final_results_without_msg.fillna(0)
final_results_without_msg = final_results_without_msg.clip(lower=-2, upper=1)
reg_values = df_secundario["Reg"][:400].clip(lower=-2, upper=1)

assert len(final_results_without_msg) == len(reg_values), "Os tamanhos n√£o coincidem!"

comparison_results = {}

for column in final_results_without_msg.columns:
    abs_diff = abs(final_results_without_msg[column] - reg_values)
    
    mae = abs_diff.mean()
    
    mse = mean_squared_error(reg_values, final_results_without_msg[column])
    
    rmse = np.sqrt(mse)
    
    comparison_results[column] = {
        "MAE": mae,      
        "MSE": mse,       
        "RMSE": rmse      
    }

comparison_df = pd.DataFrame(comparison_results).T

print(comparison_df)
```
